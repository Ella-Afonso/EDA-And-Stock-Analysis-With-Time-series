## Time resampling 

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='YE')
```

```python
df_tesla.resample(rule='YE').min()
```

```python
df_tesla.resample(rule='YE').max()
```

```python
## year and frequency
df_tesla.resample(rule='YE').max()['Open'].plot()
```

```python
## Quaterly Start Frequency
df_tesla.resample(rule='QS').max()
```

```python
## Quaterly Start Frequency
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
## Business End Frequency
df_tesla.resample(rule='BYE').max()['High'].plot()
```

```python
## Business End Frequency
df_tesla.resample(rule='BYE').max()
```

```python
## Business End Frequency
df_tesla.resample(rule='BQS').max()
```

```python
## Plotting

df_tesla['Open'].resample(rule='YE').mean().plot(kind='bar')
```

```python
## Mean
df_tesla['Open'].resample(rule='ME').mean()
```

```python
## Montly max

df_tesla['Open'].resample(rule='M').max()
```

```python
## Plotting Monthly Max 

df_tesla['Open'].resample(rule='M').max().plot(kind='bar')
```

```python
## Plotting Monthly Max 

df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize=(16,6))
```
